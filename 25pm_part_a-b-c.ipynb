Problem: Predict House Price

```python
import pandas as pd

data = {
    'area_sqft': [800, 1000, 1200, 1500, 1800],
    'bedrooms': [2, 2, 3, 3, 4],
    'age': [20, 15, 10, 5, 2],
    'price': [3000000, 4000000, 5000000, 6500000, 8000000]
}

df = pd.DataFrame(data)

X = df[['area_sqft', 'bedrooms', 'age']]
y = df['price']

print(X)
print(y)
```

>Input (X):

1. area_sqft: size of house

2. bedrooms: number of rooms

3. age: age of property

>Output (y):

1. price: continuous value (numeric)

2. Type: Regression -> because output is continuous

>Use Case

1. Real estate price prediction
2. Salary prediction
3. Stock price forecasting

Ans 1: Machine learning can be broadly categorized into three main types: supervised learning, unsupervised learning, and reinforcement learning.

1. In supervised learning, the model is trained on labeled data, meaning each input is paired with a correct output. The objective is to learn a mapping from inputs to outputs so that the model can make accurate predictions on new, unseen data. This approach is commonly used for tasks such as regression and classification.

2. Unsupervised learning, in contrast, deals with unlabeled data where no explicit target variable is provided. The model attempts to identify hidden patterns, structures, or relationships within the data. It is often used for tasks like clustering similar data points together or reducing the dimensionality of data to simplify analysis while preserving important information.

3. Reinforcement learning involves training a model through interaction with an environment, where it learns to make decisions by receiving rewards or penalties based on its actions. Over time, the model improves its strategy to maximize cumulative rewards. This type of learning is widely applied in areas such as game playing, robotics, and autonomous systems.
