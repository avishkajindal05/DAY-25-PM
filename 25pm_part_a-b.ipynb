> Regression

Problem: Predict House Price

```python
import pandas as pd

data = {
    'area_sqft': [800, 1000, 1200, 1500, 1800],
    'bedrooms': [2, 2, 3, 3, 4],
    'age': [20, 15, 10, 5, 2],
    'price': [3000000, 4000000, 5000000, 6500000, 8000000]
}

df = pd.DataFrame(data)

X = df[['area_sqft', 'bedrooms', 'age']]
y = df['price']

print(X)
print(y)
```

>Input (X):

1. area_sqft: size of house

2. bedrooms: number of rooms

3. age: age of property

>Output (y):

1. price: continuous value (numeric)

2. Type: Regression -> because output is continuous

>Use Case

1. Real estate price prediction
2. Salary prediction
3. Stock price forecasting

> Classification

Problem: Predict Whether Student Will Pass

```python
import pandas as pd

data = {
    'hours_studied': [1, 2, 3, 4, 5],
    'attendance': [50, 60, 70, 80, 90],
    'previous_score': [40, 50, 60, 70, 80],
    'pass': [0, 0, 1, 1, 1]
}

df = pd.DataFrame(data)

X = df[['hours_studied', 'attendance', 'previous_score']]
y = df['pass']

print(X)
print(y)
```

>Input (X):

1. hours_studied
2. attendance
3. previous_score

>Output (y):

1. pass -> categorical (0 or 1)

0 = Fail

1 = Pass

Type: Classification -> because output is discrete

>Use Case

1. Spam detection (spam / not spam)
2. Disease prediction (yes / no)
3. Fraud detection
